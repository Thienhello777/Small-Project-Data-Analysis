```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()
```

```python
# Best math schools in NYC (Way 1) (at least 80%)
best_math_schools = schools[schools["average_math"] >= 640][["school_name", "average_math"]].sort_values("average_math", ascending=False)
best_math_schools
```

```python
# Best math schools in NYC (Way 2) (at least 80%)
    # Avg math greater 80% * 800 (scores)
avg_math_greater_80per =schools[schools['average_math'] >= 0.8 *800].sort_index()
avg_math_greater_80per
    # Filter out best math schools in NYC
best_math_schools = avg_math_greater_80per.loc[:,['school_name','average_math']].sort_values('average_math', ascending = False)
best_math_schools
```

```python
# Top 10 school in NYC have high SAT score
  #Total SAT score
schools['total_SAT'] = schools['average_math'] + schools['average_reading'] + schools['average_writing']
schools['total_SAT']
  # Sort index
school_index =schools.sort_index()
school_index
  #Sort SAT score DESC
top_10_SAT = school_index.sort_values('total_SAT', ascending = False)
  #Filter 10 schools have high scores  
top_10_schools = top_10_SAT.head(10).loc[:,['school_name','total_SAT']]
top_10_schools
```

```python
# The highest Standard Deviation in NYC borough
    # Create aggreration 
borough = schools.groupby('borough')['total_SAT'].agg(['count', 'mean', 'std']).round(2)
    # Filter the highest std
largest_std_dev = borough[borough['std'] == borough['std'].max()]
    # Rename columns 
largest_std_dev = largest_std_dev.rename(columns={"count": "num_schools", "mean": "average_SAT", "std": "std_SAT"})
    # Result
largest_std_dev.reset_index(inplace=True)
largest_std_dev
```
